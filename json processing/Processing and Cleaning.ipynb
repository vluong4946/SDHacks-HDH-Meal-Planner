```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from os import listdir
from os.path import isfile, join
import ijson
```

```python
def sublist_check_from_left(sublist,lister):
    if len(lister) < len(sublist):
        return False
    
    check = True
    for i in range(0,len(sublist)):
        if sublist[i] != lister[i]:
            check = False
    return check
```

```python
leaf_set = set()
prefix_set = set()
name = "64 Degrees"
filename = name + ".json"
with open(filename, 'r') as f:
    parser = ijson.parse(f)
    count = 0
    for prefix, event, value in parser:
        prefix_set.add(prefix)
        is_leaf = True
        to_remove = []
        count += 1
        if count % 10000 == 0:
            print(count)
        for x in leaf_set:
            if sublist_check_from_left(prefix.split('.'),x.split('.')):
                is_leaf = False
            if sublist_check_from_left(x.split('.'),prefix.split('.')):
                to_remove.append(x)
        if is_leaf:
            leaf_set.add(prefix)
        leaf_set.difference_update(to_remove)
    print(count)
```

```python
leaves_list = list(leaf_set)
menus_vm = [x for x in leaf_set if x.startswith("MenusVm") and not x.endswith("DV")]
print(menus_vm)
```

```python
def removekey(d, key):
    r = dict(d)
    del r[key]
    return r

path = r"C:\Users\acer\Desktop\programming\python\SDHacks\json processing\json"
jsonfiles = [f for f in listdir(path) if isfile(join(path,f))]
dfs = []
for fn in jsonfiles:
    with open(join(path,fn), 'r') as f:
        objects = ijson.items(f, 'MenusVm.menuitems.item')
        columns_start = list(objects)
        columns_wo_nutrition = [removekey(x,'nutrition') for x in columns_start]
        nutrition_columns = [x.pop('nutrition') for x in columns_start]
        all_columns_merged = [{**columns_wo_nutrition[i],**nutrition_columns[i]} for i in range(0,len(nutrition_columns))]
        df = pd.DataFrame(all_columns_merged)
        df = df.drop([x for x in list(df) if x.endswith("DV")],axis=1)
        dfs.append(df)
```

```python
dfs_proc = []
for x in dfs:
    x['cholesterol'] = x['cholesterol'].map(lambda x: x.rstrip('mg'))
    x['dietaryFiber'] = x['cholesterol'].map(lambda x: x.rstrip('g'))
    x['protein'] = x['protein'].map(lambda x: x.rstrip('g'))
    x['saturatedFat'] = x['saturatedFat'].map(lambda x: x.rstrip('g'))
    x['sodium'] = x['sodium'].map(lambda x: x.rstrip('mg'))
    x['sugars'] = x['sugars'].map(lambda x: str(x).rstrip('g'))
    x['totalCarbohydrate'] = x['totalCarbohydrate'].map(lambda x: x.rstrip('g'))
    x['transFat'] = x['transFat'].map(lambda x: str(x).rstrip('g'))
    x['price'] = x['price'].map(lambda x: x.lstrip('$'))
    x['tags'] = x['tags'].map(lambda y: y.split(',')[0])
    x = x.rename(index=str, columns={"cholesterol": "cholesterol(mg)",
                                   "dietaryFiber": "dietaryFiber(g)",
                                   'protein': 'protein(g)',
                                   'saturatedFat': 'saturatedFat(g)',
                                   'sodium': 'sodium(mg)',
                                   'sugars': 'sugars(g)',
                                   'totalCarbohydrate': 'totalCarbohydrate(g)',
                                   'transFat': 'transFat(g)',
                                   'totalFat': 'totalFat(g)',
                                   'price': 'price($)',
                                   'tags': 'mealType'})
    x = x.drop(columns=['disclaimer', 'images', ])
    dfs_proc.append(x)
```
